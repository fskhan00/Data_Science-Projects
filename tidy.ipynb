# Pew Data

```python
pew = pd.read_csv('./data/pew.csv')
```

```python
pew
```

```python
pew_long = pd.melt(pew, id_vars='religion')
```

```python
pew_long
```

```python
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count' )
```

```python
pew_long[pew_long['income']=="Don't know/refused"] = '1'
```

```python
pew_long[pew_long['income']==">150k"] = "151k"
```

```python
variable_split= pew_long['income'].str.split('-')
```

```python
variable_split.str.get(0)
```

```python
variable_split.str.get(1)
```

```python
pew_long['min'] = variable_split.str.get(0)
```

```python
pew_long['max'] = variable_split.str.get(1)
```

```python
pew_long
```

```python
pew_long.isna().info()
```

```python
pew_long
```

```python
pew_long['max'] = pew_long['max'].fillna(pew_long['min'])
```
